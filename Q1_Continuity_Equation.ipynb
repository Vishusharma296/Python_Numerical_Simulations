## Continuity Equation ##



 $$ \frac{\partial\Sigma}{\partial t} + \nabla \cdot (\Sigma \mathbf{v}) = 0  $$
 
 Where $ \frac{\partial\Sigma}{\partial t}  $ ---- the rate of change of surface mass density\
 $ \nabla \cdot (\Sigma \mathbf{v}) $ --- the divergence of the mass flux vector. 
 
 
### Simulating for 1-D Grid using Finite Difference Scheme ###

```python
# Importing libraries 

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants and parameters

L = 100.0  # Length of the domain
N = 1000  # Number of grid points
dx = L / N  # Grid spacing
dt = 0.01  # Time step
```

```python
# Initial conditions

Sigma = np.zeros(N)
Sigma[N//2] = 1.0  # Initial Gaussian profile
```

```python
# Time evolution

for t in range(N):
    
    # Calculating the spatial derivative
    dSigma_dt = np.gradient(Sigma, dx)

    # Updating Sigma using the continuity equation
    Sigma = Sigma - dt * dSigma_dt
```

```python
# Plotting the result

x = np.linspace(0, L, N)
plt.plot(Sigma,x)
plt.title('Accretion Disk - Continuity Equation')
plt.xlabel('Surface Density (Σ)')
plt.ylabel('Position')

plt.show()
```
